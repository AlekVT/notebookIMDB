```python
import csv
from pyspark.sql import SQLContext
from pyspark.sql.types import *
from datetime import datetime

csv.register_dialect(
    'myDialect',
    delimiter = ',',
    lineterminator = '\r\n',
    skipinitialspace = False,
    quotechar = '"'
)

def readFile(fileName):
    filecsv=[]
    with open(fileName) as csvfile:
        filereader = csv.reader(csvfile, dialect='myDialect')
        for row in filereader:
            filecsv.append(row)
            #print(row)
    return filecsv

def makeDic(file):
    key = ""
    values = []
    dFilms = {}
    keys = file[0]
    file.pop(0)
    for i in range(len(file)):
        values = file[i]
        for a, b in zip(keys, values):
            if a in dFilms:
                dFilms[a] = dFilms[a] + '", "' + b
            else:                
                dFilms[a] =  b
            #print a + '\t' + b
    return dFilms

def getColumn(file, columnHeader):
    column = []
    cutstr = ""
    for i in range(len(file[0])):
        if file[0][i] == columnHeader:
            n = i
            
    for i in range(len(file)):       
        for j in range(len(file[i])):
            if(j == n):
                cutstr = '\xc2'                
                line = file[i][j]
                if cutstr in line:
                    line = line[:line.rfind(cutstr)]
                    column.append(line)
                else:
                    column.append(line)
                # print file[i][j]
    return column

def getTitleDir(file, title, director):
    row = []
    cutstr = ""
    dir = ""
    tit = ""
    for i in range(len(file[0])):
        if file[0][i] == title:
            t = i
     
    for i in range(len(file[0])):
        if file[0][i] == director:
            d = i
            
    for i in range(len(file)):       
        for j in range(len(file[i])):
            cutstr = '\xc2'                
            line = file[i][j]
            if(j == d):
                dir = line
            elif(j == t):                
                if cutstr in line:
                    tit = line[:line.rfind(cutstr)]
                    
                else:
                    tit = line
                    
                dirTit = tit, dir
                row.append(dirTit)   
    return row

file = readFile('movie_metadata.csv')
directors = getColumn(file, 'director_name')
#directors
titles = getColumn(file, 'movie_title')
#titles
actors1 = getColumn(file, 'actor_1_name')
#actors1
titlesDir = getTitleDir(file, 'movie_title', 'director_name')
#titlesDir

dic = makeDic(file)
dic['director_name']
```
